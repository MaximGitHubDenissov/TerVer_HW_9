Задача 2 Посчитать коэффициент линейной регрессии при заработной плате (zp), используя
градиентный спуск (без intercept).

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
x1 = np.array([35, 45, 190, 200, 40, 70, 54, 150, 120, 110])
y1 = np.array([401, 574, 874, 919, 459, 739, 653, 902, 746, 832])
```

```python
def mse (B1, y=y1, x =x1,n=len(x1)):
    return np.sum((B1*x-y)**2)/n

alpha = 1e-6
n = len(x1)
B1 = 0.1
```

```python
for i in range(1000):
    B1 -= alpha * (2/n)*np.sum((B1*x1-y1)*x1)
    print('B1 = {B1}, mse = {mse}'.format(B1=B1, mse = mse(B1)))
```
